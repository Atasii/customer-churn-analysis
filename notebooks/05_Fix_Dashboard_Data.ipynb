```python
# Cell 1: Fix Dashboard Data
import pandas as pd

# Load your cleaned data
df = pd.read_csv('../outputs/cleaned_data.csv')

# Load original data to get categorical columns
df_original = pd.read_csv('../data/WA_Fn-UseC_-Telco-Customer-Churn.csv')

# Add back the original categorical columns to dashboard data
df['Contract'] = df_original['Contract']
df['PaymentMethod'] = df_original['PaymentMethod']
df['InternetService'] = df_original['InternetService']
df['gender'] = df_original['gender']

# Check if columns were added
print("✅ Columns added successfully!")
print(f"New columns: Contract, PaymentMethod, InternetService, gender")
print(f"\nData shape: {df.shape}")
print(f"\nFirst few rows of new columns:")
print(df[['Contract', 'PaymentMethod', 'InternetService', 'gender']].head())

# Save
df.to_csv('../dashboard/dashboard_data.csv', index=False)
print("\n✅ Fixed dashboard data saved to: dashboard/dashboard_data.csv")
```
