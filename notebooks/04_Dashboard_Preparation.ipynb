```python
# Cell 1: Prepare data for dashboard
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Load cleaned data
df = pd.read_csv('../outputs/cleaned_data.csv')

# Load predictions from modeling
# If you saved predictions, load them. If not, we'll create them
try:
    high_risk = pd.read_csv('../outputs/high_risk_customers.csv')
    print("✅ Loaded high-risk customers")
except:
    print("⚠️ Creating sample predictions...")
    # Simple rule-based for dashboard demo
    df['churn_probability'] = df['MonthlyCharges'] / 150 + (1 - df['tenure']/100)
    df['churn_probability'] = df['churn_probability'].clip(0, 1)
    high_risk = df[df['churn_probability'] > 0.6].copy()

print(f"📊 Dashboard data ready: {len(df)} customers")
print(f"🎯 High-risk customers: {len(high_risk)}")

# Save dashboard data
dashboard_data = df.copy()
dashboard_data.to_csv('../dashboard/dashboard_data.csv', index=False)
print("✅ Dashboard data saved to dashboard/dashboard_data.csv")
```
